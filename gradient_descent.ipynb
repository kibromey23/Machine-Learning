```python
import numpy as np
import matplotlib.pyplot as plt

# Function and its gradient
def f(x): return x**2
def grad_f(x): return 2 * x

# Gradient descent types
def batch_gradient_descent(x_init, lr, steps):
    x = x_init
    trajectory = [x]
    for _ in range(steps):
        gradient = grad_f(x)  # Use all "data"
        x = x - lr * gradient
        trajectory.append(x)
    return trajectory

def stochastic_gradient_descent(x_init, lr, steps):
    x = x_init
    trajectory = [x]
    for _ in range(steps):
        # Simulate randomness (add noise like random data variation)
        noise = np.random.randn() * 0.1
        gradient = grad_f(x + noise)
        x = x - lr * gradient
        trajectory.append(x)
    return trajectory

def mini_batch_gradient_descent(x_init, lr, steps, batch_size=4):
    x = x_init
    trajectory = [x]
    for _ in range(steps):
        # Simulate a small batch by averaging random noisy gradients
        gradients = [grad_f(x + np.random.randn() * 0.1) for _ in range(batch_size)]
        avg_gradient = np.mean(gradients)
        x = x - lr * avg_gradient
        trajectory.append(x)
    return trajectory

# Run all methods
x_start = 10  # Starting point
learning_rate = 0.1
epochs = 50

traj_batch = batch_gradient_descent(x_start, learning_rate, epochs)
traj_sgd = stochastic_gradient_descent(x_start, learning_rate, epochs)
traj_mini = mini_batch_gradient_descent(x_start, learning_rate, epochs)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(traj_batch, label='Batch Gradient Descent')
plt.plot(traj_sgd, label='Stochastic Gradient Descent')
plt.plot(traj_mini, label='Mini-Batch Gradient Descent')
plt.axhline(0, color='gray', linestyle='--')
plt.xlabel('Iteration')
plt.ylabel('x value')
plt.title('Gradient Descent Types on f(x) = x²')
plt.legend()
plt.grid(True)
plt.show()
```
