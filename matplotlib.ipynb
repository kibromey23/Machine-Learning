```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
```

### Line plot

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
print("====================================================")
print(y)
print("====================================================")
print(z)
```

```python
plt.plot(x,y)
plt.plot(x,z ,'g x')
plt.title("sin wave")
plt.xlabel("xlabel")
plt.ylabel("ylabel")
plt.show()
```

```python
plt.plot(x,z ,'g x')
plt.show
```

### Bar chart

```python
subject = ["eng","math","phy","chem","bio","ict"]
marks = [90,94,67,47,79,100]
plt.grid(True)
plt.bar(subject,marks)
plt.show()
```

### Piechart

```python
plt.pie(marks,labels = subject,autopct ="%1.2f%%")
plt.show
```

### 3D visual

```python
from mpl_toolkits.mplot3d import Axes3D

# Sample student data
df = pd.DataFrame({
    'name': ['Liya', 'Kibrom', 'Mekdes', 'Samuel', 'Saron', 'Nahom', 'Ruth', 'Bereket', 'Marta', 'Yonatan'],
    'marks': [87, 92, 78, 85, 90, 75, 88, 82, 95, 80],
    'age': [21, 22, 20, 23, 21, 22, 20, 24, 22, 23],
    'department': ['CS', 'IT', 'CS', 'IS', 'CS', 'IT', 'IS', 'IS', 'CS', 'IT']
})

# Convert department to numbers
df['dep_code'] = df['department'].astype('category').cat.codes

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(df['age'], df['marks'], df['dep_code'], c=df['dep_code'], cmap='plasma')

ax.set_xlabel('Age')
ax.set_ylabel('Marks')
ax.set_zlabel('Department (code)')
ax.set_title('3D Student Data Visualization')

plt.show()
```
