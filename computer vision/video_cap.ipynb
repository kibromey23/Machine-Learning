#### working the video capture

### using for loop

```python
cap = cv2.VideoCapture(os.path.join('videos','rm.mp4'))

while True:
    ret, frame = cap.read()
    if not ret:  # stop if no more frames
        break

    cv2.imshow('Video Player', frame)

    if cv2.waitKey(25) & 0xFF == ord('q'):  # 25ms per frame ~ 40 FPS
        break

cap.release()
cv2.destroyAllWindows()
```

### using for loop

```python
#establish video capture
cap = cv2.VideoCapture(os.path.join('videos','rm.mp4')) 
#looping through each frame 
for frame_idx in range (int(cap.get(cv2.CAP_PROP_FRAME_COUNT))):
    #read frame 
    ret,frame = cap.read() 

    #changing into grey color
    grey = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    #show image 
    if cv2.waitKey(25) & 0xFF == ord('q'):  # 25ms per frame ~ 40 FPS
        break
    cv2.imshow('Video Player',grey) 
cap.release() 
cv2.destroyAllWindows()
```
