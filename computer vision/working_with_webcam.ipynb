#### 1.Importing dependecies

```python
import cv2
import matplotlib.pyplot as plt
```

#### 2.Rendering in real time

```python
#connect to webcam
cap = cv2.VideoCapture(3)
ret,frame = cap.read()

#loop through every frame till we close our webcam
while cap.isOpened():
    ret , frame = cap.read()

    if not ret:   # ✅ important fix
        break

    #show image
    cv2.imshow('webcam' , frame)

    #checks whether q has been hit and loop stops
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

#release te webcam 
cap.release()

#close the frame
cv2.destroyAllWindows()
    
```
