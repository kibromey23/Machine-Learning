### 1.Impoting dependecies

```python
import cv2
import matplotlib.pyplot as plt
import os
```

#### 2. Load image 

```python
img = cv2.imread('images/bk.jpeg')
## displaying the image in BGR
plt.imshow(img)
```

```python
## changing the color into RGB
rgb = cv2.cvtColor(img , cv2.COLOR_BGR2RGB)
plt.imshow(rgb)
```

```python
#resize image
resized = cv2.resize(img , (int(img.shape[1]/2),int(img.shape[0]/2)))
#view image using opencv
cv2.imshow('Frame View',resized)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 3. Edge detection

```python
canny = cv2.Canny(img ,threshold1=200,threshold2 =210)
```

```python
resized = cv2.resize(canny , (int(img.shape[1]*2),int(img.shape[0]*2)))
#view image using opencv
cv2.imshow('Frame View',resized)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 4 Export the edges

```python
cv2.imwrite('edges.jpg',canny)
```

```python
im = cv2.imread('edges.jpg')
plt.imshow(im)
```
