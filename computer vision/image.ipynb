### 1 . Importing the dependencies

```python
import cv2
import os 
import matplotlib.pyplot as plt
%matplotlib inline
```

### 2. Read in image

```python
img_path = os.path.join('images','kb.jpeg')
```

```python
img_path
```

```python
img = cv2.imread(img_path)
```

```python
img.shape
```

```python
height,width,channels = img.shape
```

```python
height
```

```python
plt.imshow(img)
```

### 3. opencv reads the image is BGR so we need to recolor into RGB 

```python
recolor = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(recolor)
```

#### changing into Gray color

```python
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(gray)
```

#### write out images

```python
cv2.imwrite('gray.jpeg',gray)
print("image saved successfully")
```

### Reread the saved image

```python
read_gray = cv2.imread('gray.jpeg')
plt.imshow(read_gray)
```
