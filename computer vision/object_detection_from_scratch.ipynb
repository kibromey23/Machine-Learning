#### Importing dependencies

```python
import os
import uuid
import time
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
img_path = os.path.join('images','captured')
labels = ['awake','drowsy']
img_nums = 20
```

```python
cap = cv2.VideoCapture("http://192.168.137.75:4747/videofeed")

#loop through labels
for label in labels:
    print(f'collecting images for {label}')

    # 5 seconde time delay to runn for the next frame
    time.sleep(5)

    #loop through img_nums
    for img_num in range (img_nums):
        print(f'collecting images for {label} image number {img_num}')

        #webcam feed
        ret,frame = cap.read()

        #naming
        img_name = os.path.join(img_path,label +'.'+str(uuid.uuid1())+'.jpg')

        #write out image to file
        cv2.imwrite(img_name,frame)

        #image rendering to the screen
        cv2.imshow('frame collection',frame)

        #2 seconde time delay to runn for the next frame
        time.sleep(2)

        if cv2.waitKey(10)&0xFF ==ord('q'):
            break
cap.release()
cv2.destroyAllWindows()
```
