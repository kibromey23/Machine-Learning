## label encoding means changing categorical class in to numerical values
### Label Encoding is used to convert categorical labels into numeric values

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
from sklearn.datasets import  load_iris
```

```python
iris = load_iris()
```

```python
df = pd.DataFrame(data = iris.data ,columns = iris.feature_names )
df['target'] = iris.target
```

```python
df.head()
```

```python
df["target"].value_counts()
```

### this is already encoded and i want to import unencoded dataset and i will encode myself

```python
df2 = pd.read_csv('iris.csv')
```

```python
df2.head()
```

```python
df2['Species'].value_counts()
```

#### now we need to encode the Spesies Class 

```python
le = LabelEncoder()
df2['encoded'] = le.fit_transform(df2['Species'])
```

```python
df2.head()
```

```python
df2.tail()
```

## they are gonna encoded based on alphbate
#### 0 ---> Setosa
#### 1 ---> Versicolor
#### 2 ---> Virginica

#### now lets remove the categorical class which is Species clss

```python
df2.drop("Species",axis = 1, inplace=True)
```

```python
df2.head()
```

### lets decode back to original using inverse_transform()

```python
df2["Species"] = le.inverse_transform(df2['encoded'])
```

```python
df2.head()
```

```python
df2.tail()
```

### if we want to remove it we just use df.drop('column_name',axis=1,inplace= True)

```python
df2.drop('encoded',axis=1,inplace=True)
```
