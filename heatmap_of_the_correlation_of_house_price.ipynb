```python
from scipy.stats import poisson
import matplotlib.pyplot as plt

#  = 3 expected events per interval
x = range(0, 10)
pmf = poisson.pmf(x, mu=3)

plt.bar(x, pmf)
plt.title("Poisson Distribution (=3)")
plt.xlabel("Number of Events (k)")
plt.ylabel("Probability")
plt.show()
```
