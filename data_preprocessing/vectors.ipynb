**Dot product and cross products**

**Dot Product** = How similar two directions are?
                  Measures cosine similarity between vectors (used in NLP, embeddings, etc.)


**Cross Product** = What is a vector that's perpendicular to two given vectors?
      Used in computer graphics and robotics for calculating rotations, normals.


```python
# Define two vectors
a = np.array([3, 4])
b = np.array([2, 1])

# Dot Product
dot = np.dot(a, b)
print("Dot Product:", dot)

# Define two 3D vectors for cross product
a3 = np.array([1, 2, 3])
b3 = np.array([4, 5, 6])

# Cross Product
cross = np.cross(a3, b3)
print("Cross Product:", cross)
```

**projection of vectors**

Vector Projection is about "shadowing" one vector onto another vector.

 Imagine you have two vectors, a and b.

 The projection of a onto b is the part of a that points in the same direction as b.

 It's like dropping a perpendicular line from the tip of a onto b, and seeing how far along b it lands.

 It's super important in Machine Learning for understanding similarity, feature reduction, and optimization.

```python
# Vectors
a = np.array([3, 4])
b = np.array([2, 0])

# Dot product
dot_product = np.dot(a, b)

# Magnitude squared of b
b_magnitude_squared = np.dot(b, b)

# Projection
projection = (dot_product / b_magnitude_squared) * b

print("Projection of a onto b:", projection)
```
