```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("creditcard.csv")
df.head()
```

```python
df["Class"].value_counts()
```

### 0 ---> legit
### 1 ---> fruad
#### this dataset is highly imbalanced so we need to handle this

#### 1. we need to separate the legit and ilegit(fruad)

```python
legit = df[df['Class'] ==0]
```

```python
fruad = df[df["Class"]==1]
```

```python
legit.shape
```

```python
fruad.shape
```

### <font color =green> we have 492 fruad so lets take 492 sample legit transaction </font>

#### 2. we use undersampling to handle the imbalanced dataset
#### our main goal is to build a sample dataset containing similar distribution of legit and fruadlent transactions

```python
legit_sample = legit.sample(n = 492)
```

```python
legit_sample.shape
```

### Now lets conctnate the legit_sample and fruad transacions to new df which is " handled_df "

```python
handled_df = pd.concat([legit_sample,fruad],axis=0)
```

```python
handled_df.shape
```

```python
handled_df.head()
```

```python
handled_df.tail()
```
