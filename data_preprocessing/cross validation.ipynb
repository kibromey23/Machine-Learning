### spliting the data and target or label

```python
X = data.drop(columns=['target'],axis=1)
y=data['target']
```

### train test split

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,stratify=y,random_state=3)
```

```python
print(X_train.shape,X_test.shape,y_train.shape,y_test.shape)
```

### compairing the models

```python
#list of models
models =[LogisticRegression(max_iter = 5000) , SVC(kernel='linear'),KNeighborsClassifier(),RandomForestClassifier()]
```

```python
def compare_models():
    for model in models:
        model.fit(X_train,y_train)
        prediction=model.predict(X_test)
        accuracy = accuracy_score(prediction,y_test)
        print(f'accuracy of {model} : {accuracy} ')
```

```python
compare_models()
```

### cross validation
cross validation of logistic regression

```python
cross_val = cross_val_score(LogisticRegression(max_iter=5000),X,y,cv=5)
```

```python
cross_val_mean = sum(cross_val)/len(cross_val)*100
round(cross_val_mean,1)
```

#### cross validation of all models

```python
def cross_val_models():
    for model in models:
        cross_val_all = cross_val_score(model,X,y,cv=5)
        print(f" cross validation of {model}:{cross_val_all}")
        cross_mean = sum(cross_val_all)/len(cross_val_all)*100
        print(f'accuracy of {model}: {round(cross_mean,2)}')
        
    
    
```

```python
cross_val_models()
```
