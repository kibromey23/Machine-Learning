### Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import seaborn as sns
```

### Data collection and Data processing

```python
#loading the dataset into pandas dataframe
data = pd.read_csv("sonar data.csv",header=None)
# why header = None ??? becuase pandas use the first row as header if we haven't put header to None
```

```python
#to see the first 5 top rows
data.head()
```

```python
#to see the last 5 rows
data.tail()
```

```python
#to see how many rows and columns are there in the dataset
data.shape
```

```python
#to see the statistical representation of the dataset
data.describe()
```

```python
#to view the information of the dataset
data.info()
```

```python
# to check how many values in each class
data[60].value_counts()
```

```python
#separating data and labels
X = data.drop(columns=60,axis=1)
y=data[60]
```

```python
X.head()
```

```python
y.head()
```

#### train test split

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.1 ,stratify=y,random_state=1)
```

```python
print("===X_train====")
print(X_train.head())
print("===X_test====")
print(X_test.head())
print("===y_train====")
print(y_train.head())
print("===y_test====")
print(y_test.head())
```

### Model Training

```python
model = LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

### model evaluation

```python
#accuracy on training data
prediction = model.predict(X_train)
accuracy = accuracy_score(prediction,y_train)
print("Accuracy on test data : ",accuracy)
```

### Model evaluation using confusion matrix

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns

cf_matrix = confusion_matrix(y_train,prediction)
cf_matrix
```

```python
tn,fp,fn,tp=  cf_matrix.ravel()
print(tn,fp,fn,tp)
```

### visualizing using seaborn

```python
sns.heatmap(cf_matrix,annot=True)
```

```python
#accuracy on test data
prediction_test = model.predict(X_test)
accuracy=accuracy_score(prediction_test,y_test)
print("Accuracy on test data : ", accuracy)
```
