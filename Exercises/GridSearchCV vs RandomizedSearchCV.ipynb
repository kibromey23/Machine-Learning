### Importing dependencis

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
from sklearn.svm import SVC
```

```python
# Loading the dataset 
breast_data = load_breast_cancer()
```

```python
# looking what the dataset looks like
breast_data
```

```python
#loading the dataset into pandas dataframe and storing into data(features) and labels(targets)
X= pd.DataFrame(breast_data.data,columns=breast_data.feature_names)
y=pd.DataFrame(breast_data.target)
```

```python
#displaying the first top 5 rows of the dataset
X.head()
```

```python
# dispalying the target or the labels
y
```

### 1 -- malignant ==== defective
### 0 -- benign === healthy

```python
# to check how many of in each class
y.value_counts()
```

```python
X =np.asarray(X)
y =np.asarray(y)
```

### Loading the model

```python
model = SVC()
```

#### Hyperparametrs

```python
parametrs = {
    "kernel":['linear','poly','rbf','sigmoid'],
    "C":[1,5,10,20]
}
```

###  **1. Grid search**

```python
classifier = GridSearchCV(model,parametrs,cv=5)
```

#### Fitting the model

```python
classifier.fit(X,y)
```

```python
result = pd.DataFrame(classifier.cv_results_)
```

```python
result
```

#### best parametrs

```python
classifier.best_params_
```

#### so the best parametrs are kernel = linear , c=10

#### highest accuracy

```python
classifier.best_score_
```
