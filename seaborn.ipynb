```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
iris = sns.load_dataset("iris")
iris.head()
```

```python
iris["species"].value_counts()
```

### filtering the species 

```python
setosa = iris[iris["species"]=="setosa"]
```

```python
setosa
```

```python
versicolor = iris[iris["species"]=="versicolor"]
```

```python
versicolor
```

```python
virginica = iris[iris["species"]=="virginica"]
```

```python
virginica
```

### visualizing the species 

```python
sns.scatterplot(x="sepal_length",y="petal_length" ,hue= "species", data=iris)
plt.show()
```

```python
sns.scatterplot(x="sepal_length" , y="sepal_width",hue="species", data=iris)
plt.show()
```

```python
sns.scatterplot(x="sepal_width" , y="petal_width" ,hue="species" , data=iris)
plt.show()
```

### counting categoris with countplot()

```python
sns.countplot(x="species" , data =iris)
plt.show()
```

### ploting bar to see mean or sum of the speicies categorically

#### to visualize sum of each categories in species use estimator = np.sum

```python
sns.barplot(x="species" ,y="sepal_width",hue="species",estimator=np.sum,data = iris)
plt.show()
```
